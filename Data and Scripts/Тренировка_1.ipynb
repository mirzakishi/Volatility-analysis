# The relativedelta type is designed to be applied to an existing datetime and can replace specific components of that datetime, or represents an interval  of time. There are two different ways to build a relativedelta instance. The

1.first one is passing it two date/datetime classes:
    
      relativedelta(datetime1, datetime2)

2.The second one is passing it any number of the following keyword arguments:
                
      relativedelta(arg1=x,arg2=y,arg3=z...)





```python
from datetime import datetime
from dateutil.relativedelta import relativedelta, MO

dt = datetime(2018, 4, 9, 13, 37, 0)
delta = relativedelta(hours=25, day=1, weekday=MO(1))
dt + delta

#First, the day is set to 1 (the first of the month), 
#then 25 hours are added, to get to the 2nd day and 14th hour, 
#finally the weekday is applied, but since the 2nd is already a Monday there is no effect



```

```python

#Relative days

# day and days 
from dateutil.relativedelta import relativedelta
from datetime import date
dt=date.today()  # today is 2019-09-23

# Adding one day to today 
print(dt + relativedelta(days=1))  # 2019-09-24

# 1st day of the month 
print(dt + relativedelta(day=1))  # 2019-09-01

```

```python
#Python dates
#The strftime() method returns a string representing date and time using date, time or datetime object.

from datetime import datetime
now = datetime.now() # current date and time

year = now.strftime("%Y")
print( year)

month = now.strftime("%m")
print("month:", month)

day = now.strftime("%d")
print("day:", day)

time = now.strftime("%H:%M:%S")
print("time:", time)

date_time = now.strftime("%Y-%m-%d, %H:%M:%S")
print("date and time:",date_time)
```

```python
#Creating string from a timestamp

from datetime import datetime

timestamp = 1528797322
date_time = datetime.fromtimestamp(timestamp)
print("Output 1",date_time)

d = date_time.strftime("%m/%d/%Y, %H:%M:%S")
print("Output 2:", d)

d = date_time.strftime("%d %b, %Y")
print("Output 3:", d)

d = date_time.strftime("%d %B, %Y")
print("Output 4:", d)

d = date_time.strftime("%I%p")
print("Output 5:", d)
```

# Python For Loops

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.



```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
#Looping Through a String

#Even strings are iterable objects, they contain a sequence of characters:

for x in "banana":
  print(x)
```

```python
#The break Statement

#With the break statement we can stop the loop before it has looped through all the items:

#After print
#fruits = ["apple", "banana", "cherry"]
#for x in fruits:
 # print(x) 
  #if x == "banana":
   # break

#Before print
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

```python
#With the continue statement we can stop the current iteration of the loop, and continue with the next:

fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```
